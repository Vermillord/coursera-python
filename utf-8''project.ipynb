```python
import zipfile

from PIL import Image
import pytesseract
import cv2 as cv
import numpy as np

# loading the face detection classifier
face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')

# the rest is up to you!
```

# Global dictionary and unzip function

```python
# Global dict and list for images and files
small_img = {}
images = {}
# Dict format img_name:[image, text in image]
file_list_small = []
file_list = []

def unzip(zip_, img_, f_list):
    zf = zipfile.ZipFile(zip_)
    
    for i in zf.infolist():
        img_[i.filename] = [Image.open(zf.open(i.filename))]
        f_list.append(i.filename)
```

## Small file preprocess

```python
unzip('readonly/small_img.zip', small_img, file_list_small)
```

```python
for file in file_list_small:
    img = small_img[file][0]
    #display(img)
    
    small_img[file].append(pytesseract.image_to_string(img).replace('-\n',''))
```

## Large file preprocess

```python
unzip('readonly/images.zip', images, file_list)
```

```python
for file in file_list:
    img = images[file][0]
    #display(img)
    
    images[file].append(pytesseract.image_to_string(img).replace('-\n',''))
```

# Search Function

```python
def search(text, image_dict, f_list):
    for file in f_list:
        img = image_dict[file][0]
        
        # Check if the word is in the file
        if text in image_dict[file][1]:
            print('Results found in file', file)
            # Try to find faces
            try:
                faces = (face_cascade.detectMultiScale(np.array(img),1.35,4)).tolist()
                # Add faces to the dictionary
                image_dict[file].append(faces)
                
                faces_in_file = []
                
                for x,y,w,h in image_dict[file][2]:
                    # Modify image
                    faces_in_file.append(img.crop((x,y,x+w,y+h)))
                
                # Create contact sheet to show images found
                contact_sheet = Image.new(img.mode, (550,110*int(np.ceil(len(faces_in_file)/5))))
                
                x = 0
                y = 0
                
                for face in faces_in_file:
                    face.thumbnail((110,110))
                    # Show faces
                    contact_sheet.paste(face, (x, y))
                    
                    if x+110 == contact_sheet.width:
                        x=0
                        y=y+110
                    else:
                        x=x+110
                
                display(contact_sheet)

            except:
                print('But there were no faces in that file!')
```
